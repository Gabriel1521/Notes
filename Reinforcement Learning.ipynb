# Reinforcement Learning

## 1. Introduction 

1.1.Characteristics of Reinforcement Learning
+ No supervisor, just reward signal
+ Feedback delayed, not instantaneous
+ Time matters (sequential data)
+ Agent action affect subsequence data received

1.2.Examples of Reinforcement Learning

+ Fly stunt manoeuvres in a helicopter
+ Defeat the world champion at Backgammon
+ Manage an investment portfolio
+ Control a power station
+ Make a humanoid robot walk
+ Play many different Atari games better than human

1.3.Reward
+ A reward Rt is a scalar feed back signal
+ Indicates how well agent is doing at step t
+ Agent's job is to maximise cumulative reward

Definition  
All goals can be described by the maximisation of expected cumulative reward

1.4.Sequential Decision Making
+ Goal: select actions to maximise total future reward
+ Actions may have long term consequences
+ Reward may be delayed
+ It may be better to sacrifice immediate reward to gain more long-term reward

1.5.History and State
+ History is the sequence of observations, actions, rewards



+ State is the information used to determine what happens next


+ What happens next depends on the history:
 + agent selects actions
 + environment selects observations/rewards

1.6.Information State

Information state(a.k.a. Markov state) contains all useful information from the history.


+ The future is independent of the past given the present



+ Once state is know, the history may be thrown away
+ state is sufficient statistic of the future

1.7.Fully Observable Environments  
Full observability: agent directly observes environment state


+ Agent state = environment state = information state
+ Formally, this is a Markov decision process

1.8.Partially Observable Environments
+ Partial observability: agent indirectly observes environment:
 + A robot with camera vision isn't toalk its absolute location
 + A poker playing agent only observes public cards
+ Partially observable Markov decision process(POMDP)


1.9.Major Components of RL Agent
+ RL agent may include one or more of these components:
    + Policy: agent's behaviour function
    + Value function: how good is each state and/or action
    + Model: agent's representation of the environment

2.0.Policy

+ Policy is the agent's behaviour
+ It is a map from state to action
+ Deterministic policy: a = pi(s)
+ Stochastic policy pi(a|s) = P[At=a|St=s]

2.1.Value Function
+ Value function is a prediction of future reward
+ Used to evaluate the goodness/badness of states
+ Therefore to select between actions


2.2.Model
+ A model predicts what the environment will do next
+ P predicts the next state
+ R predicts the next (immediate) reward



2.3.Learning and Planning
+ Two fundamental problems in sequential decision making
 + Reinforcement learning
  + The environment is initially unknown
  + The agent interacts with the environment
  + The agent improves its policy
 + Planning
  + A model of the environment is known
  + The agent performs computations with its model
  + The agent improves its policy

2.4.Exploration and Exploitation
+ Exploration finds more information about the environment
+ Exploitation exploits know information to maximise reward

2.5.Prediction and Control
+ Prediction: evaluate the future
 + Given a policy
+ Control: optimise the future
 + Find the best policy
